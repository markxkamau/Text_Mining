```python
import nltk
import numpy as np
import string
import matplotlib.pyplot as plt
```

```python
import pandas as pd
from nltk.tokenize import word_tokenize

# declaring dataset we are to work with
dataset = pd.read_csv('./covid19_tweets.csv')
docs = dataset["text"]

# declaring number of tweets to be 
tweet_length = 10000

# Selecting upto "set_length" tweets from the dataset
data = docs[:tweet_length]

# Remove anything but characters and spaces
valid_data = data.str.replace('[^A-z ]','').str.replace(' +',' ').str.strip()

# Empty token variable to transform the valid data into a single dimension
token =[]

# loop to add the valid data into the token
for i in range(tweet_length):
    token.append(valid_data[i]) 

# convert the tweets to lower case
lower_tweet_tokens = [word.lower() for word in word_tokenize(str(token))] 

# Output on the data 
print("Data: ",valid_data[5],"\n----------------------------------------\nTweet tokens: ",lower_tweet_tokens,"\n")
```

```python
from nltk.text import Text
t = Text(lower_tweet_tokens)
t.concordance("today")
```

```python
import re
womaen_strings=[w for w in lower_tweet_tokens if re.search('m[ae]ny', w)]
print(womaen_strings)
```

```python
from nltk.probability import FreqDist
fdist = FreqDist(lower_tweet_tokens)
fdist
```

```python
# total number of word count in the corpus
fdist.N()
# We can also determine the relative frequency of a token in a corpus, so what % of the corpus a term is:
fdist.freq('covid')
# And count the number of times a token appears in a corpus:
fdist['covid']
```

```python
# Visualising Frequency distributions of tokens in text

fdist.plot(30,title='Frequency distribution for 30 most common tokens in our text collection')
```

```python
# Removing Stop Words

from nltk.corpus import stopwords
remove_these = set(stopwords.words('english') + list(string.punctuation) + list(string.digits))
filtered_text = [w for w in lower_tweet_tokens if not w in remove_these]
fdist_filtered = FreqDist(filtered_text)
fdist_filtered.plot(30,title='Frequency distribution for 30 most common tokens in our text collection (excluding stopwords and punctuation)')
```

```python
# Word Cloud
from collections import Counter
dictionary=Counter(filtered_text)
import matplotlib.pyplot as plt
from wordcloud import WordCloud

cloud = WordCloud(max_font_size=80,colormap="hsv").generate_from_frequencies(dictionary)
plt.figure(figsize=(16,12))
plt.imshow(cloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
